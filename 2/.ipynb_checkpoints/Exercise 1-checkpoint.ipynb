```python
m = df1.population.values.size
plt.title('Profit as a function of population in 10,000s')
plt.scatter(df1.population.values, df1.profit.values, marker='x', color='red', linewidth=0.5)
plt.show()
```

```python
# Compute the loss value
def computeCost(X, Y, theta):
    return np.sum(np.square(X.dot(theta) - Y) * 1/(2 * Y.shape[0]))
```

```python
X = np.zeros((m, 2))
theta = np.zeros((2,1))
X[:, 0] = 1
X[:, 1] = np.array(df1.population.values)
Y = np.array(df1.profit.values).reshape((m, 1))
```

```python
computeCost(X, Y, np.zeros((2,1)))
```

```python
computeCost(X, Y, np.array([-1, 2]).reshape((2, 1)))
```

```python
def gradientDescent(X, Y, theta, iterations = 1500, alpha = 0.01):
    history = [0] * iterations
    t_history = [0] * iterations
    m = Y.shape[0]
    for i in range(iterations):
        history[i] = computeCost(X, Y, theta)
        t_history[i] = theta
        theta = theta - (alpha * 1 / m * np.sum(( X.dot(theta) - Y ) * X, axis = 0)).reshape(theta.shape)
    return (theta, history, t_history)
```

```python
res = gradientDescent(X, Y, np.zeros((2, 1)))

plt.plot(res[1][2:])
plt.title('Cost of as a function of iterations')
plt.show()
```

```python
theta_hat = res[0]
hist = np.array(res[1])
t_hist = np.array(res[2])

X.dot(theta_hat)
plt.plot(X[:,1], X.dot(theta_hat))
plt.scatter(df1.population.values, df1.profit.values, marker='x', color='red', linewidth=0.5)
plt.title('Linear regression line plotted in the scatter')
plt.show()
```

```python
B0 = np.linspace(-10, 10, 50)
B1 = np.linspace(-1, 4, 50)
xx, yy = np.meshgrid(B0, B1, indexing='xy')
Z = np.zeros((B0.size,B1.size))

for (i,j), v in np.ndenumerate(Z):
    Z[i,j] = computeCost(X,Y, np.array([xx[i,j], yy[i,j]]).reshape((2,1)))

    
Z = Z
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

CS = ax1.contour(xx, yy, Z, np.logspace(-2, 3, 20), cmap=plt.cm.jet)
ax1.scatter(theta_hat[0],theta_hat[1], marker='x', c='r')

ax2.plot_surface(xx, yy, Z, rstride=1, cstride=1, alpha=0.8, cmap=plt.cm.jet)
ax2.set_zlabel('Cost')
ax2.set_zlim(Z.min(),Z.max())
ax2.view_init(elev=15, azim=230)
```
